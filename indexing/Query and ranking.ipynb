```python
from collections import Counter,defaultdict
from bs4 import BeautifulSoup
import os
import glob
import nltk
import math
from __future__ import division
import re
import string
import pickle
import csv

def index_docs(query, invertedIndex):
    invertedDocIndex = defaultdict(Counter)
    hits = []
    
    for term in process_input(query):
        for doc, freq in invertedIndex[term].items():
                invertedDocIndex[doc][term] += invertedIndex[term][doc]
    return invertedDocIndex

def process_input(query):
    return [w for w in [w.strip(string.punctuation) for w in nltk.word_tokenize(query.lower())] if w != '']
```

```python
def tfidf(term, document):
    """
    Given a single query term and a document, score the document using
    tf-idf.
    """
    
    invertedDocIndex = index_docs(term,invertedIndex)
    #print(invertedDocIndex)
    try:
        # term frequency in document
        TFtd =  invertedDocIndex[document][term]

        # document frequency in which term comes forward
        DFt =  len(invertedIndex[term])

        # total number of docs in total
        N = amount_docs

        return round((1 + math.log10(TFtd)) * math.log10(N / DFt), 4)
    
    except:
        return 0.0
```

```python
def rankedSearchTfidf(query):
    """
    Given a query, score the documents according to the tf-idf scoring
    scheme and print the documents and scores in decreasing order.
    """ 
    
    invertedDocIndex = index_docs(query,invertedIndex)
    processed_input = process_input(query)
    
    scores = {}
    for doc,index in invertedDocIndex.items():
        score = 0.0
        for term in processed_input: 
            score += tfidf(term, doc)
            
        scores[doc] = round(score, 4)
    return sorted(scores.items(), key=lambda x: x[1], reverse=True)[:10]
```

```python
import numpy


def normalizedVector(actors_dict):
    """
    Given a document, vectorize and normalize it
    """
    
    all_vectors_docs = {}
    for actor, terms in actors_dict.items():
        vectorized = [tfidf(term, actor) for term in terms]
        divider = numpy.sqrt(sum([x*x for x in vectorized]))
        normalized = [x/divider for x in vectorized]
        all_vectors_docs[actor] =  numpy.array(normalized)
    
    return all_vectors_docs
```

```python
def cosineSimilarity(query,docVectorNormalized):
    """
    Given a query and a document, compute and print the cosine-similarity
    between the two.
    """
    invertedDocIndex = index_docs(query,invertedIndex)
    processed_input = process_input(query)
    scores = {}
    
    for doc,index in invertedDocIndex.items():
        queryVector = [tfidf(term, doc) for term in processed_input]
        queryDivider = numpy.sqrt(sum([x*x for x in queryVector]))
        queryVectorNormalized = [x/queryDivider for x in queryVector] 
        sumxx = 0
        sumyy = 0
        sumxy = 0
        
        for i in range(len(queryVectorNormalized)):
            
            x = queryVectorNormalized[i]; y = docVectorNormalized[doc][i]
            sumxx += x*x
            sumyy += y*y
            sumxy += x*y
            
        cosSim = sumxy/math.sqrt(sumxx*sumyy)
        scores[doc] = cosSim
    
    return sorted(scores.items(), key=lambda x: x[1], reverse=True)[:10]
```
